```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
path = 'site-logs/session-logs.csv'
data = pd.read_csv(path, header=None, names=['username', 'date', 'time', 'action'])
data
```

```python
data['username'].unique().shape[0]
```

```python
data['action'].unique()
```

```python
# finding out the distinct set of users and the number of times they have logged out in the data
d1 = data[['username', 'action']]
d1
```

```python
s1 = d1['action'] == 'out'
s2 = s1.loc[s1 == True]
d2 = d1.loc[s2.index]
d2
```

```python
d3 = d2['username']
d3
```

```python
d4 = d3.value_counts()
d4
```

```python
d4.plot(kind='bar')
plt.tight_layout()
plt.savefig('logouts_count.pdf', format='pdf')
```

```python
# finding out the distinct set of users along with the number of times that they have logged out as well as logged in
d1
```

```python
d5 = d1.copy()
d5['total'] = 1
d5
```

```python
d6 = d5.groupby(['username', 'action'])
d6
```

```python
d7 = d6.sum()
d7
```

```python
d8 = d7.unstack()
d8
```

```python
print(d8.index)
print(d8['total', 'out'])
```

```python
d8.plot(kind='bar')
```

```python
d9 = d5.pivot_table('total', aggfunc='sum', index='username', columns=['action'])
d9
```
