```python
import pandas as pd
import numpy as np
from pandas import DataFrame
from json import loads
```

```python
path = 'url-shortner/usagov_bitly_data2012-03-16-1331923249.txt'
file = open(path, mode='r')
lines = []
for line in file:
    lines.append(loads(line))
```

```python
data = DataFrame(lines)
data
```

```python
s1 = data['tz']
s2 = s1.fillna('Missing')
s2[s2 == ''] = 'Missing'
s2
```

```python
data['tz'] = s2
data.head(20)
```

```python
top_10_tz = data['tz'].value_counts().head(10)
top_10_tz
```

```python
top_10_tz.plot(kind='barh')
```

```python
data['os'] = np.where(data['a'].str.contains('Windows'), 'Windows', 'Non-Windows')
data[['a', 'os']]
```

```python
d1 = data[['tz', 'os']]
d1
```

```python
d1['t'] = 1
d1
```

```python
d2 = d1.pivot_table('t', aggfunc='sum', index='tz', columns=['os'])
d3 = d2.fillna(0)
d3['total'] = d3['Non-Windows'] + d3['Windows']
```

```python
d4 = d3.sort_values(by='total', ascending=False).head(10)
del d4['total']
d4
```

```python
d4.plot(kind='bar')
```
